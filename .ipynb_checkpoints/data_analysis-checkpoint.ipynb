```python
import pandas as pd
import numpy as np
import os
import sys
import shutil
import re
import json
import matplotlib.pyplot as plt  
import seaborn as sns
from IPython.display import Markdown as md
```

```python
def parse_time_str(s: str) -> list[float]:
    pattern = r":(\d+\.?\d*)"
    numbers = re.findall(pattern, s)
    # convert to float
    result = [float(num) for num in numbers]
    return result
def parse_time_series(s):
    # convert to float
    return np.array(np.array(parse_time_str(s), dtype=np.float64), dtype=np.float64)
```

```python
df = pd.read_csv('result.csv')
df.head()
```

```python
valid_cnt = df[df["valid_proj"] == True]['valid_proj'].sum()
md(f"#### 使用 FFI Analyzer 得发现使用到FFI函数的crate有{valid_cnt}个")
```

```python
md(f"#### 使用 FFI Analyzer 得发现使用到FFI函数的crate有{valid_cnt*100/build_success_cnt}%")
```

```python
valid_df = df[df['valid_proj'] == True]
normal_build_times = valid_df["normal_build_time"].apply(parse_time_series)
valid_normal_build_times = valid_df["normal_build_time"].apply(parse_time_series)
ffi_checker_build_times = valid_df["ffi_checker_build_time"].apply(parse_time_series)
ffi_analyzer_times = valid_df["ffi_checker_analysis_time"].apply(parse_time_series)
analysis_time_sum = ffi_analyzer_times.sum(axis=0)
build_time_sum = normal_build_times.sum(axis=0)
gen_ir_time_sum = ffi_checker_build_times.sum(axis=0)
average_addition_cost = ((gen_ir_time_sum - build_time_sum) / build_time_sum) * 100
md(f"""
使用FFI Checker功能额外带来的编译时间开销为:\n
real_time: {average_addition_cost[0] *100 //1 /100}% \n
user_time: {average_addition_cost[1] *100 //1 /100}% \n
sys_time: {average_addition_cost[2] *100 //1 /100}%""")
```

```python
average_analysis_cost = (analysis_time_sum / build_time_sum) * 100
md(f"""
使用FFI Analyzer 功能额外带来的编译时间开销为:\n
real_time: {average_analysis_cost[0] *100 //1 /100}%\n
user_time: {average_analysis_cost[1] *100 //1 /100}%\n
sys_time: {average_analysis_cost[2] *100 //1 /100}%""")
```

```python
ffi_checker_build_times = np.array([np.array(x) for x in ffi_checker_build_times])
normal_build_times = np.array([np.array(x) for x in normal_build_times])
ffi_analyzer_times = np.array([np.array(x) for x in ffi_analyzer_times])

addition_time_cost = (ffi_checker_build_times - normal_build_times) / normal_build_times
analyzer_addition_time_cost = ffi_analyzer_times / normal_build_times
aatc_data = analyzer_addition_time_cost
atc_data = addition_time_cost

figure1, axis_list = plt.subplots(3, 2)
figure1.set_size_inches(10, 8)
plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9)
print(axis_list)
sns.histplot(atc_data[:,0], bins=20, kde=True, ax=axis_list[0][0]).set_title("Real time additional time cost distribution diagram")
sns.histplot(atc_data[:, 1], bins=20, kde=True, ax=axis_list[1][0]).set_title("User time additional time cost distribution diagram")
sns.histplot(atc_data[:, 2], bins=20, kde=True, ax=axis_list[2][0]).set_title("System time additional time cost distribution diagram")

sns.histplot(aatc_data[:,0], bins=20, kde=True, ax=axis_list[0][1]).set_title("Real time additional time cost distribution diagram")
sns.histplot(aatc_data[:, 1], bins=20, kde=True, ax=axis_list[1][1]).set_title("User time additional time cost distribution diagram")
sns.histplot(aatc_data[:, 2], bins=20, kde=True, ax=axis_list[2][1]).set_title("System time additional time cost distribution diagram")
figure1.tight_layout(h_pad=2, w_pad=1)
figure1.show()
figure1.savefig("./figures/figure1.eps", format="eps", dpi=300, bbox_inches="tight")
figure1.savefig("./figures/figure1.svg", format="svg", dpi=300, bbox_inches="tight")
```

```python
def is_costly(row):
    
    nbt = parse_time_series(row.normal_build_time)
    cbt = parse_time_series(row.ffi_checker_build_time)
    if nbt[0] * 2 < cbt[0]:
        return True
    else:
        return False

costly_rows = []
for row in valid_df.itertuples(index=False):
    if is_costly(row):
        costly_rows.append(row)
costly_df = pd.DataFrame(costly_rows)
costly_df[["name","normal_build_time", "ffi_checker_build_time", "ffi_checker_analysis_time"]].head()
```

```python
normal_build_times = costly_df["normal_build_time"].apply(parse_time_series)
normal_build_times = np.array([np.array(x) for x in normal_build_times])
costly_series = pd.Series(normal_build_times[:,0])
print(costly_series.describe())
figure2, axis = plt.subplots(1,1)
# axis = axis[0]
axis = sns.histplot(costly_series, bins=20)
axis.set_title("Costly crate build time distribution diagram")
axis.set_xlabel("Second")
axis.set_ylabel("Count")
figure2.savefig("./figures/figure2.svg", format="svg", dpi=300)
figure2.savefig("./figures/figure2.eps", format="eps", dpi=300)
```

```python
# ffi_data = {"ffi_name": [], "ffi_width":[], "ffi_depth":[], "ffi_usage_rate": []}
crate_info_data = {"crate_name": [], "ffi_width_mean": [], "ffi_depth_mean":[], "ffi_usage_rate": []}
for row in valid_df.itertuples(index=False):
    dirname = row.dirname
    result_path = os.path.join(os.getcwd(), "result_collect", dirname)
    interface_str = ""
    with open(os.path.join(result_path, "interface.json")) as f:
        interface_str = f.read()
    interface_info = json.loads(interface_str)
    ffi_cnt = len(interface_info)
    ffi_width = []
    ffi_depth = []
    for ffi_info in interface_info:
        # print(ffi_info)
        ffi_width.append(len(ffi_info["call_stack"]))
        ffi_depth.append(np.array([len(call_stack) for call_stack in ffi_info["call_stack"]]).max())
    crate_ffi_width_mean = np.array(ffi_width).mean()
    crate_ffi_depth_mean = np.array(ffi_depth).mean()
    usage_rate = (np.array(ffi_depth) != 0).sum() / ffi_cnt

    crate_info_data["crate_name"].append(dirname)
    crate_info_data["ffi_width_mean"].append(crate_ffi_width_mean)
    crate_info_data["ffi_depth_mean"].append(crate_ffi_depth_mean)
    crate_info_data["ffi_usage_rate"].append(usage_rate)
```

```python
crate_info_df = pd.DataFrame(crate_info_data)
crate_info_df.head()
```

```python
figure3, axis3_list = plt.subplots(1, 3)
print(axis3_list.shape)
figure3.set_size_inches(15, 5)
plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9)
sns.histplot(crate_info_df["ffi_width_mean"], bins=20, kde=True, ax=axis3_list[0]).set_title("FFI Width Mean")
sns.histplot(crate_info_df["ffi_depth_mean"], bins=20, kde=True, ax=axis3_list[1]).set_title("FFI Depth Mean")
sns.histplot(crate_info_df["ffi_usage_rate"], bins=20, kde=True, ax=axis3_list[2]).set_title("FFI Usage Rate")
figure3.tight_layout(h_pad=2, w_pad=1)
figure3.show()
figure3.savefig("./figures/figure3.eps", format="eps", dpi=300, bbox_inches="tight")
figure3.savefig("./figures/figure3.svg", format="svg", dpi=300, bbox_inches="tight")
```
